# Deep Switching Autoregresive Factorization (DSARF)

DSARF package is an easy-to-use implementation of our [AAAI'21 paper](https://arxiv.org/pdf/2009.05135.pdf) which can be used for training, testing, and visulization in just few lines of code. In this notebook, we show how to use DSARF for short- and long-term prediction and state estimation on several vector time series datasets.

```python
from dsarf import DSARF
import numpy as np
import torch
torch.manual_seed(10)
np.random.seed(10)
```

## Example 1: Bat Flight Dataset

The dataset is a list of $N=10$ skeletal data of size $(T=166-436)\times (D=34\times 3)$. We keep the last two data for test. We downsample data in time by a factor of 5.

```python
import json
path = "./data/bat.json"
f = open(path)
data = json.load(f)
D = len(data['joints']) * 3
data = [np.array(data[key]).reshape(-1, D)[::5]
           for key in list(data.keys())[3:-2]]
data_train = data[:-2]
data_test = data[-2:]
```

After loading the dataset, we initilize the DSARF model with arguments `D: feature dimension of data`, `factor_dim: latent dimension`, `L: list of temporal lags`, `S: number of discrete states`. We set `batch_size = 1` as this dataset contains data with different temporal length. Alternatively, we could pad all the data with Nans to the same length and use a higher batch size.

```python
dsarf = DSARF(D, factor_dim =5, L=[1,2], S=2, batch_size=1, recursive_state = True)
```

Now, we are ready to fit the train data on the model for 300 epochs. The model outputs ELBO loss each epoch, and reconstruction error and next-time-point prediction error every 50 epochs.

```python
model_train = dsarf.fit(data_train, epoch_num = 300)
```

Once the model is trained, we can run inference on a test set.

```python
model_test = dsarf.infer(data_test)
```

We can use the built-in function in the output model to plot states for few of the data instances (around 10 instances sampled at fixed intervals). We can access inferred states by calling `model.q_s` which returns a tensor of size $N\times T_{max}\times S$. 

```python
model_train.plot_states()
model_test.plot_states()
```

We can get the short-term predicted data by calling `y_pred, y_pred_n, y_pred_p = model.short_predict()`
which returns predicted data and their $\pm\sigma$ uncertainty intervals. We can simply plot short-term predicted vs groundtruth time series (for few intances and dimensions) by passing the grountruth data as input to the built-in function `model.plot_predict(data)` or call `model.report_stats(data)` to obtain reconstruction and short-term prediction error.

```python
model_test.plot_predict(data_test)
```

```python
#salsa
torch.manual_seed(10)
np.random.seed(10)

root_dir = "./data/salsa.json"

import json
f = open(root_dir)
data_st = json.load(f)
D = len(data_st['joints']) * 3
data_st = [np.array(data_st[key]).reshape(-1, D)[::20]
           for key in list(data_st.keys())[1:]]
                
dsarf = DSARF(D, factor_dim =5, L=[1,2], S=3, batch_size=1, recursive_state = True)#, recurrent=True)

model_train = dsarf.fit(data_st[:-1], 300)
model_test = dsarf.infer(data_st[-1:], 500)
model_train.plot_states()
model_test.plot_states()
model_test.plot_predict(data_st[-1:])
```

```python
#birmingham

torch.manual_seed(10)
np.random.seed(10)

root_dir = "./data/birmingham.mat"

from scipy.io import loadmat
data_st = loadmat(root_dir)['tensor'].transpose(1,2,0).astype('float')
data_st[data_st == 0] = np.nan

D = data_st.shape[-1]
                
dsarf = DSARF(D, factor_dim =10, L=[1,2], S=1, batch_size=30)#, recurrent=True)

model_train = dsarf.fit(data_st[:-7], 500)
model_test = dsarf.infer(data_st[-7:], 500)
model_test.plot_predict(data_st[-7:])
```

```python
#birmingham

torch.manual_seed(10)
np.random.seed(10)

root_dir = "./data/birmingham.mat"

from scipy.io import loadmat
data_st = loadmat(root_dir)['tensor'].transpose(1,2,0).astype('float')
data_st[data_st == 0] = np.nan

D = data_st.shape[-1]
T = data_st.shape[1]
data_st = data_st.reshape(1,-1,D)
                
dsarf = DSARF(D, factor_dim =10, L=[1,2,3,18,19,20,126,127,128], S=1, batch_size=1)#, recurrent=True)

model_train = dsarf.fit(data_st[:,:-7*T], 500)
model_train.plot_predict(data_st[:,-7*T:], 7*T)
```

```python
y_pred, _, _ = model_train.long_predict(7*T)
idxs = ~np.isnan(data_st[:,-7*T:])
np.sqrt(((y_pred - data_st[:,-7*T:])[idxs]**2).mean())/np.sqrt((data_st[:,-7*T:][idxs]**2).mean())*100
```

```python
# apnea

torch.manual_seed(10)
np.random.seed(10)

root_dir = "./data/apnea.txt"

import pandas as pd
data_st = [pd.read_csv(root_dir, sep=' ', header = None).values[6201:7201,1].reshape(-1,1),
            pd.read_csv(root_dir, sep=' ', header = None).values[5201:6201,1].reshape(-1,1)]

D = 1
                
dsarf = DSARF(D, factor_dim =1, L=[1], S=2, batch_size=1, recursive_state=False)# recurrent=True)

model_train = dsarf.fit(data_st[:-1], 2000)
model_test = dsarf.infer(data_st[-1:], 500)
model_train.plot_states()
model_test.plot_states()
model_test.plot_predict(data_st[-1:])
```
